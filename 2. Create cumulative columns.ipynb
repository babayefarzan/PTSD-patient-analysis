```python
import pandas as pd
import numpy as np
```

```python
data_df = pd.read_csv('reduced.csv')
```

```python
#remove HR column as it is irrelevant
data_df.index = np.arange(1,len(data_df)+1)
data_df.drop(['HR'], inplace=True,axis=1)
```

```python
#add Nan count column that has count of total Nans in a row
data_df['Nan_count'] = data_df.isnull().sum(axis=1)

#add symptom sum column which is the sum of all non-Nan values in a row instance
#if nan values -1
#data_df.replace(-1,0,inplace=True)
data_df['symptom_sum'] =  data_df['Reexp1']+ data_df['Reexp2']+data_df['Avoid1']+ data_df['Avoid2']+ data_df['NACM1']+ data_df['NACM2']+ data_df['AAR1']+ data_df['AAR2']+ data_df['Sleep']+ data_df['Pain']
```

```python
#Adding a new column with the cumulative average of values: reexp1 till Pain
data_df.fillna(value=0, inplace=True)
data_df['Reexp1_cum'] =  data_df['Reexp1'].cumsum()
data_df['Reexp2_cum'] =  data_df['Reexp2'].cumsum()
data_df['Avoid1_cum'] =  data_df['Avoid1'].cumsum()
data_df['Avoid2_cum'] =  data_df['Avoid2'].cumsum()
data_df['NACM1_cum'] =  data_df['NACM1'].cumsum()
data_df['NACM2_cum'] =  data_df['NACM2'].cumsum()
data_df['AAR1_cum'] =  data_df['AAR1'].cumsum()
data_df['AAR2_cum'] =  data_df['AAR2'].cumsum()
data_df['Sleep_cum'] =  data_df['Sleep'].cumsum()
data_df['Pain_cum'] =  data_df['Pain'].cumsum()
```

```python
c=1;
while(c!=1144):
    i1 = data_df.get_value(c, 'Reexp1_cum', takeable=False)
    data_df.set_value(c, 'Reexp1_cum',i1/c)
    
    i2 = data_df.get_value(c, 'Reexp2_cum', takeable=False)
    data_df.set_value(c, 'Reexp2_cum',i2/c)
    
    i3 = data_df.get_value(c, 'Avoid1_cum', takeable=False)
    data_df.set_value(c, 'Avoid1_cum',i3/c)
    
    
    i4 = data_df.get_value(c, 'Avoid2_cum', takeable=False)
    data_df.set_value(c, 'Avoid2_cum',i4/c)
    
    i5 = data_df.get_value(c, 'NACM1_cum', takeable=False)
    data_df.set_value(c, 'NACM1_cum',i5/c)
    
    i6 = data_df.get_value(c, 'NACM2_cum', takeable=False)
    data_df.set_value(c, 'NACM2_cum',i6/c)
    
    i7 = data_df.get_value(c, 'AAR1_cum', takeable=False)
    data_df.set_value(c, 'AAR1_cum',i7/c)
    
    i8 = data_df.get_value(c, 'AAR2_cum', takeable=False)
    data_df.set_value(c, 'AAR2_cum',i8/c)
    
    i9 = data_df.get_value(c, 'Sleep_cum', takeable=False)
    data_df.set_value(c, 'Sleep_cum',i9/c)
    
    i10 = data_df.get_value(c, 'Pain_cum', takeable=False)
    data_df.set_value(c, 'Pain_cum',i10/c)
    
   
    c = c+1
```
